```python
import nltk
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('gutenberg')
from nltk.corpus import gutenberg
```

```python
sample = gutenberg.raw("bryant-stories.txt")
```

```python
sample = sample.lower()
```

```python
tokenized_text = sent_tokenize(sample)
```

```python
print(tokenized_text[5:15])
```

```python
len(tokenized_text)
```

```python
token_words = word_tokenize(sample)
```

```python
len(token_words)
```

```python
print(token_words[0:10])
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
capword = RegexpTokenizer('[A-Z]\w+')
```

```python
capword.tokenize(sample)
```

```python
number_token = RegexpTokenizer('[0-9]\w+')
number_token.tokenize(sample)
```

```python
from nltk.tokenize import BlanklineTokenizer
```

```python
BlanklineTokenizer().tokenize(sample)
```

```python
from nltk.probability import FreqDist
```

```python
freqdist = FreqDist(token_words)
print(freqdist)
```

```python
freqdist.most_common(20)
```

```python
freqdist.plot(20)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python
filtered_word = [word for word in token_words if word not in stop_words and len(word)>3 ]
```

```python
freq_words = FreqDist(filtered_word)
```

```python
print(freq_words)
```

```python
freq_words.most_common(20)
```

```python
from nltk import bigrams
```

```python
A = bigrams(filtered_word)
fdist = FreqDist(A)
fdist.most_common(20)
```

```python
from nltk import trigrams
```

```python
B = trigrams(filtered_word)
fdist = FreqDist(B)
fdist.most_common(20)
```

```python
from nltk import ngrams
```

```python
C = ngrams(filtered_word,5)
fdist = FreqDist(C)
fdist.most_common(20)
```

```python
fdist.plot(20)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

stem_words = [ps.stem(word) for word in filtered_word]
```

```python
stem_words[0:20]
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatized_word = [lemmatizer.lemmatize(word) for word in filtered_word]
```

```python
lemmatized_word[0:20]
```
